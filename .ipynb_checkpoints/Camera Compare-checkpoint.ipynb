# Camera Location Comparison
## Evaluation of variance in estimated external camera paramters

### first run for CUAS data collect in Austria with Bramor PPK


```python
import sys, os, glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def parseCamFile(dataDir, camFile):
    headers = pd.read_csv(dataDir + camFile, skiprows=1, sep=",").columns
    cameras = pd.read_csv(dataDir + camFile, skiprows=2, sep="\t", names=headers)
    cameras['run'] = str(camFile.split('.')[0])
    return cameras
    
    
```

```python
# make list of dataframes to refer to programatically... i.e., for each camera file make a dataframe with unique name
dataDir = "CUAS/cameras/" 
fileList = next(os.walk('CUAS/cameras'))[2] 


#camFile = fileList[5]
#print(camFile)
#print(type(camFile))
#fileList
```

```python
idx = 0
for file in fileList:
    if idx == 0:
        cam = parseCamFile(dataDir, file)
        idx += 1
    else:
        cami = parseCamFile(dataDir, file)
        cam = pd.concat([cam, cami])
        idx += 1

#cam
```

```python
cam1 = cam.reset_index()
cam1.columns
```

```python
cam1.groupby(cam1['index']).min().to_csv('CUAS_min.csv', sep=",")
```

```python
cam1.groupby(cam1['index']).max().to_csv('CUAS_max.csv', sep=",")
```

```python
cam1.groupby(cam1['index']).var().to_csv('CUAS_var.csv', sep=",")
```
